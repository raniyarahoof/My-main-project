## üîÑ **CONVERTING DATATYPES**

##### Converting to datetime allows easy date-based operations, such as: Calculating the length of stay: (Discharge Date - Date of Admission).days, Extracting year, month, or day for analysis,Sorting or filtering records by date.

```python
df['Date of Admission']= pd.to_datetime(df['Date of Admission'],dayfirst=True)
df['Discharge Date']= pd.to_datetime(df['Discharge Date'],dayfirst=True)
```

```python


# Convert numeric columns 

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Billing Amount'] = pd.to_numeric(df['Billing Amount'], errors='coerce')

# Convert repeated-value text columns into category 
category_columns = [
    'Gender', 'Blood Type', 'Medical Condition', 'Doctor',
    'Hospital', 'Insurance Provider', 'Admission Type',
    'Medication', 'Test Results'
]

for col in category_columns:
    df[col] = df[col].astype('category')


```

```python
print("DataType corrected Succesfully!!")
df.dtypes
```

## üÜï **GENERATING NEW COLUMNS**

##### To analyze patterns among different age groups instead of individual ages and to  analyze trends, e.g., which age group has more hospital visits or higher bills.

```python
# Categorize patients into age groups
def age_group(age):
    if age < 18:
        return 'Teen'
    elif age < 35:
        return 'Young Adult'
    elif age < 60:
        return 'Middle Age'
    else:
        return 'Senior'

df['Age Group'] = df['Age'].apply(age_group)
df[['Age', 'Age Group']].head()
```

##### To analyze financial analysis by grouping bills into meaningful ranges and to identify patterns in patient spending and hospital revenue.

```python
# Categorize patients based on billing amount
def billing_segment(amount):
    if amount >= 50000:
        return 'High Billing'
    elif amount >= 10000:
        return 'Medium Billing'
    else:
        return 'Low Billing'

df['Billing Segment'] = df['Billing Amount'].apply(billing_segment)

df[['Billing Amount', 'Billing Segment']].head()
```

## üìä **DATA VISUALIZATION**üìâ

##### This section presents visual representations of the dataset to uncover patterns and trends. Distributions, category comparisons, and relationships between age, billing, length of stay, and gender are analyzed to provide clear insights into patient demographics and hospital data.

### **1Ô∏è‚É£Count plot: Billing Segment Distribution**

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='Billing Segment', order=['Low Billing', 'Medium Billing', 'High Billing'], palette='cool')
plt.title('Distribution of Patients by Billing Segment', fontsize=16,fontweight='bold',color='purple')
plt.xlabel('Billing Segment', fontsize=10,fontweight='bold')
plt.ylabel('Number of Patients', fontsize=10,fontweight='bold')
plt.show()
```

#### **Observation :**
Patients are grouped into Low, Medium, and High Billing segments, giving an overview of revenue   distribution.
Most patients fall into the Low Billing category, while High Billing patients are fewer but contribute significantly to revenue.
This insight highlights where the hospital generates most of its revenue and indicates which patient groups may need focused financial strategies.

### **2Ô∏è‚É£ Count plot: Age Group Distribution** 

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='Age Group', order=['Teen', 'Young Adult', 'Middle Age', 'Senior'], palette='winter')
plt.title('Distribution of Patients by Age Group', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Age Group', fontsize=10, fontweight='bold')
plt.ylabel('Number of Patients', fontsize=10, fontweight='bold')
plt.show()
```

#### **Observation:**
Patients are grouped into Teens, Young Adults, Middle Age, and Seniors, giving a clear overview of the hospital‚Äôs patient demographics.

Young Adults and senior patients form the largest groups, while Teens and young adult are fewer in number.

This insight helps understand which age groups the hospital serves most frequently, guiding patient care planning, resource allocation, and service improvement.

### **3Ô∏è‚É£ Bar plot: Average Billing by Age Group**

```python
plt.figure(figsize=(10,8))
age_billing = df.groupby('Age Group')['Billing Amount'].mean().reindex(['Teen', 'Young Adult', 'Middle Age', 'Senior'])
age_billing.plot(kind='bar', color='plum')
plt.title('Average Billing Amount by Age Group', fontsize=16, fontweight='bold', color='darkgreen')
plt.xlabel('Age Group', fontsize=12, fontweight='bold')
plt.ylabel('Average Billing Amount', fontsize=12, fontweight='bold')
plt.xticks(rotation=0)
plt.show()
```

### **Observation:**
The average billing amount for each age group, giving insight into revenue contribution by demographics.

Seniors and Middle Age patients have higher average bills, while Teens and Young Adults have lower bills.

This suggests older patients may require costlier treatments or longer stays, helping the hospital plan financial and resource allocation strategies.

### **4Ô∏è‚É£ Bar plot: Average Length of Stay by Age Group**

```python
plt.figure(figsize=(10,8))
age_stay = df.groupby('Age Group')['Length of Stay'].mean().reindex(['Teen', 'Young Adult', 'Middle Age', 'Senior'])
age_stay.plot(kind='bar', color='turquoise')
plt.title('Average Length of Stay by Age Group', fontsize=16, fontweight='bold', color='turquoise')
plt.xlabel('Age Group', fontsize=12, fontweight='bold')
plt.ylabel('Average Length of Stay (Days)', fontsize=12, fontweight='bold')
plt.xticks(rotation=0)
plt.show()
```

#### **Observation:**
The average number of days patients stay in the hospital for each age group.

Seniors and Middle Age patients tend to stay longer, while Teens and Young Adults have shorter stays.

This insight helps the hospital plan staffing, beds, and resource allocation, and indicates that older patients may need more intensive care or treatment.

### **5Ô∏è‚É£ Pie chart: Gender Distribution**

```python
plt.figure(figsize=(10,8))
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['purple','plum'], startangle=90)
plt.title('Gender Distribution of Patients', fontsize=16, fontweight='bold', color='darkmagenta')
plt.show()
```

#### **Observation:**

 The chart reveals the proportion of male and female patients in the hospital dataset.
 A slightly higher proportion of female patients indicates that certain hospital services may attract or serve females more frequently.
The gender distribution is relatively balanced, ensuring that healthcare planning can cater effectively to both genders.
 This demographic insight can also reveal potential patterns in healthcare needs or service utilization between males and females.


### 6Ô∏è‚É£ Pie Chart: Distribution of Medical Conditions**

```python
# Pie Chart: Distribution of Medical Conditions
plt.figure(figsize=(10,8))
df['Medical Condition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    explode=[0.05]*df['Medical Condition'].nunique(),
    cmap='tab20'
)
plt.title('Distribution of Medical Conditions', fontsize=16, fontweight='bold', color='darkred')
plt.ylabel('')   
plt.show()
```

#### **Observation:**

The percentages of all conditions are very close, ranging from 12.3% to 12.7%, indicating that the patient load is evenly distributed among different medical conditions.

Flu (12.7%), Diabetes (12.6%), and Obesity (12.6%) are slightly more common, but the difference is minimal.

Other conditions like Cancer, Asthma, Heart Disease, Alzheimer‚Äôs, and Infections have shares between 12.3% and 12.5%, showing a relatively balanced prevalence.

### **7Ô∏è‚É£ Heatmap of correlations**


```python
plt.figure(figsize=(10,6))
corr = df[['Age', 'Billing Amount', 'Room Number', 'Length of Stay']].corr()  # Selecting numeric columns
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features', fontsize=16, fontweight='bold', color='darkred')
plt.show()
```

#### **Observation:**
Age and Billing Amount have a moderate positive correlation, indicating older patients tend to have higher bills.
Length of Stay shows a positive correlation with Billing Amount, meaning longer stays usually generate more revenue.
Room Number shows little correlation with other variables, suggesting room assignment is independent of billing or age.
This insight helps the hospital predict revenue based on patient age and stay duration and optimize resource allocation.

### **üîü Line Plot: Average Billing vs Length of Stay**


```python
plt.figure(figsize=(10,8))
billing_stay = df.groupby('Length of Stay')['Billing Amount'].mean().reset_index()
plt.plot(billing_stay['Length of Stay'], billing_stay['Billing Amount'], marker='o', color='darkorange', linewidth=2)
plt.title('Average Billing Amount vs Length of Stay', fontsize=16, fontweight='bold', color='orangered')
plt.xlabel('Length of Stay (Days)', fontsize=12, fontweight='bold')
plt.ylabel('Average Billing Amount', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Patients who stay longer usually have more complicated treatments, so their bills are higher.
Most patients stay for a few days, but the line rises steadily for longer stays.
This plot helps the hospital understand that longer stays mean higher revenue.
It can also help in planning beds, staff, and resources for patients who need longer care.
